```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("car_crashes")
df
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

# Scatterplot

```python
sns.scatterplot(x="ins_losses",y="speeding",data=df,hue="abbrev")
```

```python
sns.set()
fig, axes = plt.subplots(2, 2)
sns.scatterplot(y="ins_losses",x="speeding",data=df,ax=axes[0,0],color='g',legend="full")
sns.scatterplot(y="not_distracted",x="alcohol",data=df,ax=axes[0,1],color='y',legend="full")
sns.scatterplot(y="speeding",x="alcohol",data=df,ax=axes[1,0],color='pink',legend="full")
sns.scatterplot(y="total",x="no_previous",data=df,ax=axes[1,1],color='brown',legend="full")
fig.tight_layout()
```

# Lineplot

```python
sns.lineplot(x="speeding",y="ins_losses",data=df)
```

```python
sns.set()
fig, axes = plt.subplots(2, 2)
sns.lineplot(y="total",x="speeding",data=df,ax=axes[0,0],color='g')
sns.lineplot(y="total",x="alcohol",data=df,ax=axes[0,1],color='brown')
sns.lineplot(y="total",x="not_distracted",data=df,ax=axes[1,0],color='r')
sns.lineplot(y="total",x="no_previous",data=df,ax=axes[1,1],color='y')
fig.tight_layout()
```

# Histogram

```python
sns.histplot(df.total,kde=True,color='g')
```

```python
sns.set()
fig,axes=plt.subplots(2,2)
sns.histplot(df.speeding,ax=axes[0,0],kde=True,color='r')
sns.histplot(df.alcohol,ax=axes[0,1],kde=True,color='b')
sns.histplot(df.not_distracted,ax=axes[1,0],kde=True,color='y')
sns.histplot(df.no_previous,ax=axes[1,1],kde=True,color='pink')
fig.tight_layout()
```

# Piechart

```python
fig=plt.figure(figsize=(20,20))   #to zoom-in, zoom-out
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #[left,bottom,width,height]
palette_color = sns.color_palette('Set2')
axes1.pie(df.total,labels=df.abbrev,autopct="%0.1f%%",colors=palette_color)
axes1.legend()
```

# Distplot

```python
sns.distplot(df.total,color='g')
```

```python
sns.set()
fig, axes = plt.subplots(2, 2)
sns.distplot(df.speeding,ax=axes[0,0],color='y')
sns.distplot(df.alcohol,ax=axes[0,1],color='b')
sns.distplot(df.not_distracted,ax=axes[1,0],color='orange')
sns.distplot(df.no_previous,ax=axes[1,1],color='pink')
fig.tight_layout()
```

# Barplot

```python
plt.subplots(figsize=(16,10))
sns.barplot(y='total', x='abbrev', data=df)
```

# Pairplot

```python
sns.pairplot(df)
```

# Jointplot

```python
sns.jointplot(x='speeding',y='total',data=df)
```

```python
sns.jointplot(x='total',y='alcohol',data=df,kind='reg')
```

```python
sns.jointplot(x='total',y='not_distracted',data=df,kind='kde')
```

# Relplot

```python
sns.relplot(data=df, x="ins_losses", y="total",style="ins_premium")
```

```python
df1=[df.total,df.speeding,df.not_distracted,df.no_previous]
sns.relplot(data=df1,kind="line")
```

# Countplot

```python
plt.figure(figsize=(20,5))
sns.countplot(x='total',data=df,orient='h')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='alcohol',data=df,orient='h',palette="Set2")
```

# Violin plot

```python
df1=[df.total,df.speeding,df.not_distracted,df.no_previous]
sns.violinplot(df1)
plt.xticks([0,1,2,3],labels=['total','speeding','not_distracted','no_previous'])
plt.show()
```

```python
df1=[df.ins_premium,df.ins_losses]
sns.violinplot(data=df1)
plt.xticks([0,1],labels=['ins_premium','ins_losses'])
plt.show()
```

# Boxplot

```python
df1=[df.total,df.speeding,df.not_distracted,df.no_previous]
sns.boxplot(df1,palette='pastel')
plt.xticks([0,1,2,3],labels=['total','speeding','not_distracted','no_previous'])
plt.show()
```

```python
df1=[df.ins_premium,df.ins_losses]
sns.boxplot(data=df1,palette='Set2',notch=True)
plt.xticks([0,1],labels=['ins_premium','ins_losses'])
plt.show()
```

# Regression plot

```python
fig,axes=plt.subplots(2,2)
sns.regplot(x='ins_premium',y='ins_losses',data=df,ax=axes[0,0],color='b')
sns.regplot(x='speeding',y='alcohol',data=df,ax=axes[0,1],color='black')
sns.regplot(x='alcohol',y='no_previous',data=df,ax=axes[1,0],color='brown')
sns.regplot(x='not_distracted',y='total',data=df,ax=axes[1,1],color='violet')
fig.tight_layout()
```

# Pointplot

```python
plt.subplots(figsize=[5,10])
sns.pointplot(data=df,x="total",y="abbrev",dodge=True,color='orange',errorbar=("pi", 100))
```

# Swarmplot

```python
df1=[df.total,df.speeding,df.not_distracted,df.no_previous]
sns.swarmplot(data=df1)
```

```python
df1=[df.ins_premium,df.ins_losses]
sns.swarmplot(data=df1,dodge=True,palette="dark",marker='*')
```
